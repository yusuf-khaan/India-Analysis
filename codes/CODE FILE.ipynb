```python
import pandas as pd
import json
with open('gdp_1960_2020.csv') as f:
    data=pd.read_csv(f)
pd.DataFrame(data).head()
```

```python
arr=[]
for index,row in data.iterrows():
    if row['country']=='India':
        country=row['country']
        year=row['year']
        gdp=row['gdp']
        gdp_p=row['gdp_percent']
        arr.append([country,year,gdp,gdp_p])
gdp_df=pd.DataFrame(arr,columns=[country,year,gdp,gdp_p])
gdp_df.head()
```

```python
gdp_df.rename(columns={2020:'Year',2622983732006:'GDP',0.0338270131682137:'GDP%','COuntry':'Country'},inplace=True)
gdp_df
gdp_df.to_csv('india_gdp')
gdp_df.tail()
```
