```python
import pandas as pd
import json
with open('gdp_1960_2020.csv') as f:
    data=pd.read_csv(f)
    
```

```python
arr=[]
for index,row in data.iterrows():
    if row['country']=='India':
        country=row['country']
        year=row['year']
        gdp=row['gdp']
        arr.append([country,year,gdp])
gdp_df=pd.DataFrame(arr,columns=[country,year,gdp])
gdp_df.head()
```

```python
gdp_df.rename(columns = {'year':'Year','country':'Country',},inplace=True)
gdp_df
```
